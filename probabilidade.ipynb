```python
#funcao -- def + nomeFunc + par√¢metro (A, B) + :
#def cria a funcao
#.2f.format(variavel) -- mostrar duas virgulas ap√≥s a v√≠rgula.

def probab(A, E):
  resultado = (A / E)* 100
  print('{: .2f}'.format(resultado))
```

```python
#probabilidade de retirar preferencia cor preta
probab(len(cor_preta), len(dados))
```

```python
#probabilidade de de ser menor que 1.70

probab(len(altura1),len(dados))
```

```python
#probab de n√£o ocorrer um evento

def probab_nao(A, E):
  resultado = (1-(A/E))*100
  print('{: .2f}'.format(resultado))
```

```python
cor_vermelha = dados.loc[dados.cor_favorita.isin(['vermelha', 'vermelho'])]
cor_vermelha.head()
```

```python
probab_nao(len(cor_vermelha), len(dados))
```

## Probabilidade da Intersec√ß√£o de dois eventos independentes.

```python
#func da probabilidade da intersec√ß√£o de dois eventos independentes
def probab_inter(A, B, E):
  resultado = (A/E * B/E)*100
  print('{: .1f}'.format(resultado))
```

```python
#Probab de sortear duas pessoas sendo uma delas com a altura menor que 1,70
#E a outra cor favorita preta com reposi√ß√£o

probab_inter(len(cor_preta),len(altura1), len(dados))
```

##Probabilidade da Uni√£o (AUB)


```python
def probab_uniao(A, B, E):
  prob = (((A/E + B/E) - (A/E * B/E) / E) *100)
  print('{: .2f}'.format(prob))
```

```python
probab_uniao(len(cor_preta), len(altura1), len(dados))
```
